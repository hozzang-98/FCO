```python
import pandas as pd
import requests, re

from server import *
```

```python
df = pd.read_csv('data/matchdetail.csv', index_col=0)

df.head()
```

```python
func = METADATA_FUNCTION['id']

urlString = basic_link + func

response = requests.get(urlString, headers = headers)

card_info = response.json()

card_info_df = pd.DataFrame(card_info)

card_info_df.head()
```

```python
func = METADATA_FUNCTION['sid']

urlString = basic_link + func

response = requests.get(urlString, headers = headers)

season_info = response.json()

season_info_df = pd.DataFrame(season_info)

season_info_df.head()
```

```python
df['card_name'] = df['spId'].apply(lambda x:card_info_df[card_info_df['id'] == x]['name'].values[0] 
                                   if not card_info_df[card_info_df['id'] == x].empty 
                                   else "Unknown")
```

```python
df[df['card_name'] == 'Unknown']
```

```python
df = df[df['card_name'] != 'Unknown'] # 298247738
```

```python
df['season_name'] = df['spId'].apply(lambda x:season_info_df[season_info_df['seasonId'] == int(str(x)[:3])]['className'].values[0] 
                                   if not season_info_df[season_info_df['seasonId'] == int(str(x)[:3])].empty 
                                   else "Unknown")
```

```python
df[df['season_name'] == 'Unknown']
```

```python
# \s*: 괄호 앞의 공백을 제거
# \(.*?\): 괄호와 괄호 안의 모든 내용을 제거
# "": 제거된 부분을 빈 문자열로 대체
df["cleaned_season_name"] = df["season_name"].apply(lambda x: re.sub(r"\s*\(.*?\)", "", x))

df.head()
```

```python
# season_name과 card_name 결합
df["season_card"] = df["cleaned_season_name"] + " " + df["card_name"]
```

```python
df = df[['spId', 'card_name', 'season_name', 'cleaned_season_name','season_card','spPosition', 'spGrade', 'shoot', 'effectiveShoot', 'assist',
       'goal', 'dribble', 'intercept', 'defending', 'passTry', 'passSuccess',
       'dribbleTry', 'dribbleSuccess', 'ballPossesionTry',
       'ballPossesionSuccess', 'aerialTry', 'aerialSuccess', 'blockTry',
       'block', 'tackleTry', 'tackle', 'yellowCards', 'redCards', 'spRating']]
```
